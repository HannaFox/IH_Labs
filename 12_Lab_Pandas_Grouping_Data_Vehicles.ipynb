<b>show the average CO2_Emission_Grams/Km  by Brand

```python
cars_make.columns
```

Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters

```python
cars_make["CO2 Emission Grams/Mile"]=cars_make["CO2 Emission Grams/Mile"].apply(lambda x: x /1.60934)
cars_make= cars_make.rename(columns={"CO2 Emission Grams/Mile":"CO2 Emission Grams/Km"})
```

```python
cars_make = cars_make.groupby(["Make","CO2 Emission Grams/Km"]).mean()
cars_make
```

<b>show the average CO2_Emission_Grams/Km  by Brand ... sorted

```python
cars_make.sort_values(by=["CO2 Emission Grams/Km"], ascending=False)
```
